```python
import pandas as pd
import numpy as np
```

```python
class_df = pd.read_csv("class.csv")
zoo_df = pd.read_csv("zoo.csv")
```

```python
zoo_df.drop("animal_name", axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = zoo_df.drop("class_type", axis=1)
y = zoo_df["class_type"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=7, init="random", n_init=1000)

kmeans.fit(X_train)
```

```python
from sklearn.metrics import adjusted_rand_score

print( "Score on training data:", adjusted_rand_score( kmeans.predict(X_train), y_train ) )
print( "Score on testing data:", adjusted_rand_score( kmeans.predict(X_test), y_test ) )
```

```python
from sklearn.cluster import DBSCAN
from sklearn.model_selection import GridSearchCV

param_grid = {
    'eps': [0.01, 0.05, 0.1, 0.5, 1.0, 1.5, 2, 2.5, 3, 3.5],
    'min_samples': [1, 5, 10, 15, 20, 25]
}

dbscan = DBSCAN()
clf = GridSearchCV(dbscan, param_grid, scoring='accuracy')  # Use a suitable scoring metric
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_estimator_
```

```python
clf.best_score_
```

```python
best_eps = clf.best_params_['eps']
best_min_samples = clf.best_params_['min_samples']

best_dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
best_dbscan.fit(X_train)
```

```python
len(set(best_dbscan.labels_))
```

```python
from sklearn.neighbors import KNeighborsClassifier

param_grid = {
    "n_neighbors": [1, 2, 3, 4, 5, 6, 7, 8]
}

knn = KNeighborsClassifier()
clf = GridSearchCV(knn, param_grid, scoring="accuracy")
```

```python
clf.fit(X_train, y_train)
```

```python
best_clf = clf.best_estimator_
```

```python
clf.best_estimator_
```

```python
best_clf.score(X_train, y_train)
```

```python
best_clf.score(X_test, y_test)
```
