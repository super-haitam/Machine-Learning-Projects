```python
import gymnasium as gym

import torch.nn as nn
import torch

import matplotlib.pyplot as plt

import numpy as np
import random
import time
```

```python
env_name = "Pendulum-v1"
```

```python
env = gym.make(env_name, g=9.81)
```

```python
nb_states = env.observation_space.shape[0]
nb_actions = 9
```

```python
nb_states, nb_actions
```

# DDQN Class

```python
class DDQN(nn.Module):
    def __init__(self, nb_states, nb_actions):
        super().__init__()

        self.sq = nn.Sequential(
            nn.Linear(nb_states, 64),
            nn.ReLU(),
            nn.Linear(64, 64),
            nn.ReLU(),
            nn.Linear(64, nb_actions)
        )

    
        modules = self.sq.modules()
        for m in modules:
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)

    def forward(self, X: torch.Tensor):
        return self.sq(X)
```

# Agent Class

```python
class Agent:
    def __init__(self, nb_states, nb_actions, epsilon, epsilon_decay, epsilon_min, gamma, lr, max_mem_size):
        # DDQN networks
        self.online = DDQN(nb_states, nb_actions)
        self.target = DDQN(nb_states, nb_actions)
        self.copy_online_to_target()

        # Hyper-parameters
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min

        self.gamma = gamma

        # Replay Memory
        self.replay_memory = []
        self.max_mem_size = max_mem_size
        self.minibatch_size = minibatch_size

        # Loss & Optimizer
        self.loss = nn.MSELoss()
        self.optimizer = torch.optim.Adam(self.online.parameters(), lr=lr)

        self.possible_actions = list(range(nb_actions))
    
    def copy_online_to_target(self):
        self.target.load_state_dict(self.online.state_dict())

    def decrement_epsilon(self):
        self.epsilon = max(self.epsilon - self.epsilon_decay, self.epsilon_min)

    def get_action(self, state) -> int:
        # EXPLORATION
        if random.uniform(0, 1) < self.epsilon:
            action = random.choice(self.possible_actions)
        else:
            state = torch.as_tensor(state, dtype=torch.float)
            action = self.online(state).argmax().item()

        return action

    def add_to_replay_memory(self, state, action, reward, next_state, done):
        exp = {"state": state, "action": action, "reward": reward, "next_state": next_state, "done": done}

        self.replay_memory.append( exp )
        
        if len(self.replay_memory) >= self.max_mem_size:
            self.replay_memory.pop(0)

    def learn_from_replay_memory(self):
        if self.minibatch_size >= len(self.replay_memory):
            minibatch = np.random.choice(self.replay_memory, self.minibatch_size, True)
        else:
            minibatch = np.random.choice(self.replay_memory, self.minibatch_size, False)

        states =      torch.tensor( np.array(list(map(lambda x: x["state"], minibatch))), dtype=torch.float )
        actions =     torch.tensor( np.array(list(map(lambda x: x["action"], minibatch))), dtype=torch.float )
        rewards =     torch.tensor( np.array(list(map(lambda x: x["reward"], minibatch))), dtype=torch.float )
        next_states = torch.tensor( np.array(list(map(lambda x: x["next_state"], minibatch))), dtype=torch.float )
        dones =       torch.tensor( np.array(list(map(lambda x: x["done"], minibatch))), dtype=torch.int )

        return self.learn(states, actions, rewards, next_states, dones)

    def learn(self, states, actions, rewards, next_states, dones):
        self.optimizer.zero_grad()

        # Get the prediction of self.Q on states
        pred = self.online(states)
        
        # action_from_online_net is the argmax tensor of self.Q(next_states)
        _, action_from_online_net = self.online(next_states).max(1)
        
        # Get the prediction of self.target on next_states
        next_pred = self.target(next_states)

        # Since actions are one hot encoded, 'pred.mul(actions)' returns pred with only the selected actions left with the rest as 0
        pred = torch.sum(pred.mul(actions), dim=1)

        # NOTE: I have no clue what this line does, :), would be extremely nice of you if you could explain what it does and why it was
        #    used here. Thanks!
        target = rewards + (1 - dones) * gamma * next_pred.gather(1, action_from_online_net.unsqueeze(1)).squeeze(1)

        cost = self.loss(pred, target.detach())
        cost.backward()

        self.optimizer.step()
        return cost.item()
```

# Hyper Parameters

```python
epsilon = 1.0
epsilon_decay = 0.0015
epsilon_min = 0.01

gamma = 0.99
lr = 0.01

max_mem_size = 200_000
minibatch_size = 32

online_to_target_freq = 5

nb_episodes = 1_000
nb_steps = 200

verbose_freq = nb_episodes // 10
```

```python
action_to_num = [-2, -1.5, -1.0, -0.5, 0.0, 0.5, 1.0, 1.5, 2.0]
```

# Training Loop

```python
agent = Agent(nb_states, nb_actions, epsilon, epsilon_decay, epsilon_min, gamma, lr, max_mem_size)
```

```python
# Averages
episode_times = []
rep_mem_sizes = []
rewards = []
costs = []
steps = []

# Timing
action_taking_times = []
env_step_times = []
learning_times = []

for episode in range(nb_episodes):
    episode_time = time.time()
    state = env.reset()[0]

    episode_reward = 0
    episode_cost = 0

    for step in range(nb_steps):
        # Select Action
        t = time.time()
        action = agent.get_action(state)
        action_taking_times.append(time.time() - t)

        # Get s', r, d from env
        t = time.time()
        next_state, reward, done, truncated, info = env.step([action_to_num[action]])
        env_step_times.append(time.time() - t)
        
        episode_reward += reward

        # Action One Hot Encoded
        action_one_hot = np.zeros(nb_actions)
        action_one_hot[action] = 1

        # Add to replay memory
        agent.add_to_replay_memory(state, action_one_hot, reward, next_state, done)
        
        if done:
            break
    
        state = next_state

    # Copy online to target
    if episode % online_to_target_freq == 0:
            # NOTE: Time is negligeable
        agent.copy_online_to_target()

    # Learn from replay memory
    t = time.time()
    episode_cost = agent.learn_from_replay_memory()
    learning_times.append(time.time() - t)

    # Reward, Cost, Step & Rep. Mem Array
        # NOTE: Time is negligeable
    rewards.append(episode_reward)
    costs.append(episode_cost)
    steps.append(step)
    rep_mem_sizes.append(len(agent.replay_memory))

    # Decrement Epsilon
    agent.decrement_epsilon()

    # Time Elapsed Per Episode
    episode_times.append(time.time() - episode_time)

    # Verbose
    if episode % verbose_freq == 0 or episode == nb_episodes-1:
        print(f"{int((episode / nb_episodes)*100)}% | Episode: {episode} | Epsilon: {agent.epsilon:.2f} | " + \
             f"Avg. Reward: {np.mean(rewards[-verbose_freq:]):.2f} | Avg. Cost: {np.mean(costs[-verbose_freq:]):.2f} | " + \
             f"Avg. Time: {np.mean(episode_times[-verbose_freq:]):.3f}s | Rep. Mem. Size: {len(agent.replay_memory)}")

minutes = int(np.sum(episode_times) // 60)
seconds = np.sum(episode_times) % 60
print(f"\nThe training process took {minutes} minutes and {seconds:.3f}s, with an average of {np.mean(episode_times):.3f}s, " + \
      f"and a max of {np.max(episode_times):.3f}s")
```

```python
def scatter_data(data, color1, color2, padding, xlabel, ylabel, title, fig, grid, index):
    fig.add_subplot(*grid, index)
    
    plt.scatter( list(range(len(data))), data, color=color1 )

    if padding != 0:
        line = np.zeros(len(data))
        for i in range(padding, len(line)):
            line[i] = np.mean(data[i-padding: i+padding])
        line[: padding] = np.full( (padding, ), np.mean(data[: padding]) )
        
        plt.plot( list(range(len(line))), line, color=color2 )
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.title(title)
```

```python
fig = plt.figure(figsize=(18, 16))
grid = (3, 2)

scatter_data(rewards,        "C0", "C1", 100, "Episodes", "Rewards", "Evolution of Rewards by Episodes",             fig, grid, 1)
scatter_data(costs,          "C1", "C2", 100, "Episodes", "Costs", "Evolution of Costs by Episodes",                 fig, grid, 2)
scatter_data(episode_times,  "C2", "C3", 100, "Episodes", "Elapsed Time", "Evolution of Elapsed Time by Episodes",   fig, grid, 3)
scatter_data(steps,          "C3", "C4", 100, "Episodes", "Steps", "Evolution of Steps by Episodes",                 fig, grid, 4)
scatter_data(rep_mem_sizes,  "C4", "C5", 0,   "Episodes", "Rep Mem Sizes", "Evolution of Rep Mem Sizes by Episodes", fig, grid, 5)

plt.tight_layout()
plt.show()
```

```python
measured = sum(env_step_times) + sum(action_taking_times) + sum(learning_times)
actual = sum(episode_times)

print(f"{actual:.3f} - {measured:.3f} = {actual - measured:.3f}")
```

```python
times = [sum(env_step_times), sum(action_taking_times), sum(learning_times), actual - measured]
labels = [
    "Env Step",
    "Action Taking",
    "Learning From Replay Memory",
    "Others"
]

plt.pie(times, labels=labels, autopct='%1.1f%%')

plt.title(f"Time Distribution Over Training")
plt.show()
```

```python
nb_test_episodes = 1
nb_test_steps = nb_steps
```

```python
env = gym.make(env_name, render_mode="human")

for episode in range(nb_test_episodes):
    state = env.reset()[0]
    episode_reward = 0

    for step in range(nb_test_steps):
        env.render()
        
        with torch.no_grad():
            action = agent.online( torch.tensor(state) ).argmax().item()

        next_state, reward, done, truncated, info = env.step([action_to_num[action]])
        episode_reward += reward

        if done:
            break
        
        state = next_state

    print(f"Episode: {episode+1} \t| Reward: {episode_reward:.2f}")

env.close()
```
