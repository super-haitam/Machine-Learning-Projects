```python
import gym

import torch
import torch.nn as nn

import matplotlib.pyplot as plt

import numpy as np
import random
import time
```

```python
env_name = "CartPole-v1"
```

```python
env = gym.make(env_name, render_mode="rgb_array")
```

```python
nb_states = env.observation_space.shape[0]
nb_actions = env.action_space.n
```

```python
nb_states, nb_actions
```

# DQN Network Class

```python
class DQNetwork(nn.Module):
    def __init__(self, nb_states, nb_actions):
        super().__init__()

        # Sequential Model with 2 layers: INPUT & OUTPUT layers.
        # NOTE: No hidden layers here.
        self.sq = nn.Sequential(
            nn.Linear(nb_states, 128),
            nn.ReLU(),
            nn.Linear(128, nb_actions)
        )

        # Initializes its weights using Xavier (Glorot) initialization with a uniform distribution.
        # This type of initialization helps in stabilizing the training of neural networks, by setting appropriate initial
        #    values for the weights.
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)

    def forward(self, X: torch.Tensor):
        return self.sq(X)
```

# Agent Class

```python
class Agent:
    def __init__(self, nb_states, nb_actions, max_mem_size, lr, gamma, epsilon, epsilon_decay, epsilon_min):
        # Init the neural networks
        self.Q = DQNetwork(nb_states, nb_actions)
        self.target = DQNetwork(nb_states, nb_actions)
        self.copy_target_to_online()

        # Hyper Parameters
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        self.gamma = gamma

        # In Fitting the model
        self.loss = nn.MSELoss()
        self.optim = torch.optim.Adam(self.Q.parameters(), lr=lr)

        # Replay Memory
        self.replay_memory = []
        self.max_mem_size = max_mem_size
        
        self.possible_actions = list( range(nb_actions) )

    
    def copy_target_to_online(self):
        # Init the weights of self.target with the weights of self.Q
        self.target.load_state_dict(self.Q.state_dict())

    
    def get_action(self, state: torch.Tensor) -> int:
        # EXPLOITATION
        if np.random.uniform(0, 1) > self.epsilon:
            # Get the action given by the online model
            with torch.no_grad():
                action =  self.Q.forward(state).argmax().item()
        # EXPLORATION
        else:
            action = random.choice( self.possible_actions )

        return action

    
    def decrement_epsilon(self):
        # Linear Decrease
        self.epsilon = max(self.epsilon - self.epsilon_decay, self.epsilon_min)

    
    def add_to_replay_memory(self, state: int, action: int, reward: int, next_state: int, done: bool):
        # Adding the s, a, r, s', d to self.replay_memory
        self.replay_memory.append({"state": state, "action": action, "reward": reward, "next_state": next_state, "done": done})

        # Ensures that the length of self.replay_memory does not exceed self.max_mem_size
        if len(self.replay_memory) > self.max_mem_size:
            self.replay_memory.pop(0)

    
    def learn_from_replay_memory(self, minibatch_size):
        if len(self.replay_memory) < minibatch_size:
            minibatch = np.random.choice(self.replay_memory, minibatch_size, replace=True)
        else:
            minibatch = np.random.choice(self.replay_memory, minibatch_size, replace=False)
            # Replace the last element with the current observation [Optional]
            minibatch[-1] = self.replay_memory[-1]

        # Get the s, a, r, s', d from the minibatch as tensors
        states      = torch.tensor( np.array( list(map(lambda x: x["state"], minibatch)) ), dtype=torch.float)
        actions     = torch.tensor( np.array( list(map(lambda x: x["action"], minibatch)) ), dtype=torch.int)
        rewards     = torch.tensor( np.array( list(map(lambda x: x["reward"], minibatch)) ), dtype=torch.float)
        next_states = torch.tensor( np.array( list(map(lambda x: x["next_state"], minibatch)) ), dtype=torch.float)
        dones       = torch.tensor( np.array( list(map(lambda x: x["done"], minibatch)) ), dtype=torch.int)

        # Fit the model based of the s, a, r, s', d
        return self.learn(states, actions, rewards, next_states, dones)

    
    def learn(self, states: torch.Tensor, actions: torch.Tensor, rewards: torch.Tensor, next_states: torch.Tensor, dones: torch.Tensor):
        self.optim.zero_grad()

        # Get the prediction of self.Q on states
        pred = self.Q(states)
        
        # action_from_online_net is the argmax tensor of self.Q(next_states)
        _, action_from_online_net = self.Q(next_states).max(1)
        
        # Get the prediction of self.target on next_states
        next_pred = self.target(next_states)

        # Since actions are one hot encoded, 'pred.mul(actions)' returns pred with only the selected actions left with the rest as 0
        pred = torch.sum(pred.mul(actions), dim=1)

        # NOTE: I have no clue what this line does, :), would be extremely nice of you if you could explain what it does and why it was
        #    used here. Thanks!
        target = rewards + (1 - dones) * gamma * next_pred.gather(1, action_from_online_net.unsqueeze(1)).squeeze(1)

        # Compute the loss between pred and target | target.detach() cuz its grad does not matter in here.
        cost = self.loss(pred, target.detach())
        cost.backward()

        self.optim.step()

        # Return the cost for plotting and visualization
        return cost.item()
```

# Hyper Parameters

```python
max_mem_size = 100_000

lr = 0.001
gamma = 0.99

epsilon = 1.0
epsilon_decay = 0.0006
epsilon_min = 0.01

minibatch_size = 64
```

# Training Loop

```python
nb_episodes = 2_000
nb_steps = 100
```

```python
agent = Agent(nb_states, nb_actions, max_mem_size=max_mem_size, lr=lr, 
              gamma=gamma, epsilon=epsilon, epsilon_decay=epsilon_decay, 
              epsilon_min=epsilon_min)
```

```python
rewards = []
costs = []

for episode in range(nb_episodes):
    state = env.reset()[0]
    episode_rewards = 0
    episode_costs = 0

    for step in range(nb_steps):
        # Select action based on the state
        action = agent.get_action( torch.tensor(state) )
        
        # Get s', r, d after performing action
        next_state, reward, done, truncated, info = env.step(action)
        
        # Reward negatively for failing
        reward = reward if not done else -10
        
        episode_rewards += reward

        # One hot encoding the action
        action_onehot = np.zeros(nb_actions)
        action_onehot[action] = 1

        # Add all to replay memory
        agent.add_to_replay_memory(state, action_onehot, reward, next_state, done)

        if done:
            break

        state = next_state
        
    # Make the agent learn from the memory
    episode_costs = agent.learn_from_replay_memory(minibatch_size)

    # Append episode_rewards to rewards && episode_costs to costs
    rewards.append(episode_rewards)
    costs.append(episode_costs)

    # Copy online into target
    agent.copy_target_to_online()

    # Decrement epsilon
    agent.decrement_epsilon()

    # Print the evolution of some variables 10 times per training
    if episode % (nb_episodes // 10) == 0:
        print(f"{int((episode / nb_episodes) * 100)}% | Epsilon: {agent.epsilon:.3f} | Avg. Reward: {np.mean(rewards[-nb_episodes // 10:]):.1f} | " + \
              f"Avg. Cost: {np.mean(costs[-nb_episodes // 10:]):.1f} | Rep. Mem. Length: {len(agent.replay_memory)}")
```

```python
# Evolution of Rewards By Episodes
plt.scatter( list(range(len(rewards))), rewards, label="REWARD", color="C0" )

plt.xlabel("Episode")
plt.ylabel("Reward")

plt.title("Evolution of Rewards By Episodes")
plt.show()

# Evolution of Costs By Episodes
plt.scatter( list(range(len(costs))), costs, label="LOSS", color="C1" )

plt.xlabel("Episode")
plt.ylabel("Cost")

plt.title("Evolution of Costs By Episodes")
plt.show()
```
