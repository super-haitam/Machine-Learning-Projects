# Double DQN Class

```python
class DDQN(nn.Module):
    def __init__(self, nb_input, nb_output):
        super().__init__()

        self.sq = nn.Sequential(
            nn.Linear(nb_input, 128),
            nn.ReLU(),
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, nb_output)
        )

        modules = self.sq.modules()
        for m in modules:
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)

    def forward(self, X: torch.Tensor):
        return self.sq(X)
```

# Agent Class

```python
class Agent:
    def __init__(self, nb_states, nb_actions, seed, max_mem_size, minibatch_size, lr, gamma, epsilon, epsilon_decay, epsilon_min):
        torch.manual_seed(seed)
        
        # Double Deep Q Networks
        self.online = DDQN(nb_states, nb_actions)
        self.target = DDQN(nb_states, nb_actions)
        self.copy_online_to_target()

        # Hyper Parameters
        self.epsilon = epsilon
        self.epsilon_decay = epsilon_decay
        self.epsilon_min = epsilon_min
        self.gamma = gamma

        # Replay Memory
        self.replay_memory = []
        self.max_mem_size = max_mem_size
        self.minibatch_size = minibatch_size
        
        # Loss and Optimizer
        self.loss = nn.MSELoss()
        self.optim = torch.optim.Adam(self.online.parameters(), lr)

        self.possible_actions = list( range(nb_actions) )

    def get_action(self, state: np.ndarray):
        # EXPLORATION
        if np.random.uniform(0, 1) < self.epsilon:
            action = random.choice(self.possible_actions)
        # EXPLOITATION
        else:
            with torch.no_grad():
                action = self.online( torch.tensor(state) ).detach().argmax().item()

        return action

    def copy_online_to_target(self):
        self.target.load_state_dict(self.online.state_dict())

    def decrement_epsilon(self):
        self.epsilon = max(self.epsilon - self.epsilon_decay, self.epsilon_min)

    def add_to_replay_memory(self, state, action_one_hot, reward, next_state, done):
        self.replay_memory.append({"state": state, "action": action_one_hot, "reward": reward, "next_state": next_state, "done": done})

        if len(self.replay_memory) > self.max_mem_size:
            self.replay_memory.pop(0)

    def learn_from_replay_memory(self):
        if len(self.replay_memory) <= self.minibatch_size:
            return -1
        
        minibatch = np.random.choice(self.replay_memory, self.minibatch_size, False)
        minibatch[-1] = self.replay_memory[-1]
        
        states = torch.tensor( np.array(list(map(lambda x: x["state"], minibatch))), dtype=torch.float )
        actions = torch.tensor( np.array(list(map(lambda x: x["action"], minibatch))), dtype=torch.int )
        rewards = torch.tensor( np.array(list(map(lambda x: x["reward"], minibatch))), dtype=torch.float )
        next_states = torch.tensor( np.array(list(map(lambda x: x["next_state"], minibatch))), dtype=torch.float )
        dones = torch.tensor( np.array(list(map(lambda x: x["done"], minibatch))), dtype=torch.int )

        return self.learn(states, actions, rewards, next_states, dones)

    def learn(self, states, actions, rewards, next_states, dones):
        self.optim.zero_grad()

        pred = self.online(states)
        _, action_from_online_net = self.online(next_states).max(dim=1)
        next_pred = self.target(next_states)

        pred = torch.sum(pred.mul(actions), dim=1)

        target = rewards + (1 - dones) * self.gamma * next_pred.gather(1, action_from_online_net.unsqueeze(1)).squeeze(1)

        cost = self.loss(pred, target.detach())
        cost.backward()

        self.optim.step()
        return cost.item()

    def save(self, path: str):
        torch.save(self.online.state_dict(), path)
```

# Training Loop Function

```python
def train_agent(agent, hyper_params):
    # Hyper-Parameters
    online_to_target_freq = hyper_params["online_to_target_freq"]
    verbose_freq =          hyper_params["verbose_freq"]
    nb_episodes =           hyper_params["nb_episodes"]
    nb_steps =              hyper_params["nb_steps"]

    # Averages
    outcomes = np.zeros((nb_episodes, ), dtype=int)
    episode_times = []
    rep_mem_sizes = []
    epsilons = []
    rewards = []
    costs = []
    steps = []
    
    # Timing
    action_taking_times = []
    env_step_times = []
    learning_times = []
    
    print("Training Started...\n")
    for episode in range(nb_episodes):
        episode_time = time.time()
    
        state = env.reset()[0]
    
        episode_reward = 0
        episode_cost = 0

        for step in range(nb_steps):
            # Select Action
            t = time.time()
            action = agent.get_action(state)
            action_taking_times.append(time.time() - t)
    
            # Get s', r, d from env
            t = time.time()
            next_state, reward, done, truncated, info = env.step(action)
            env_step_times.append(time.time() - t)

            # Add To Wins
            if reward == WIN_REWARD:
                outcomes[episode] = 1
            elif reward == -WIN_REWARD:
                outcomes[episode] = -1

            if done and reward not in (WIN_REWARD, -WIN_REWARD): # Give Neg Reward For Not Touching The Ground
                reward = NO_LANDING_REWARD
            
            episode_reward += reward

            # Action One Hot Encoded
            action_one_hot = np.zeros(nb_actions)
            action_one_hot[action] = 1
    
            # Add to replay memory
            agent.add_to_replay_memory(state, action_one_hot, reward, next_state, done)
            
            if done:
                break
        
            state = next_state

        # Copy online to target
        if episode % online_to_target_freq == 0:
                # NOTE: Time is negligeable
            agent.copy_online_to_target()
    
        # Learn from replay memory
        t = time.time()
        episode_cost = agent.learn_from_replay_memory()
        learning_times.append(time.time() - t)
    
        # Reward, Cost, Step & Rep. Mem Array
            # NOTE: Time is negligeable
        rewards.append(episode_reward)
        costs.append(episode_cost)
        steps.append(step)
        rep_mem_sizes.append(len(agent.replay_memory))
    
        # Decrement Epsilon
        agent.decrement_epsilon()
        epsilons.append(agent.epsilon)

        # Time Elapsed Per Episode
        episode_times.append(time.time() - episode_time)

        # Verbose
        if episode % verbose_freq == 0 or episode == nb_episodes-1:
            print(f"{int((episode / nb_episodes)*100)}% | Episode: {episode} | Epsilon: {agent.epsilon:.2f} | " + \
                 f"Avg. Reward: {np.mean(rewards[-verbose_freq:]):.2f} | Avg. Cost: {np.mean(costs[-verbose_freq:]):.2f} | " + \
                 f"Avg. Time: {np.mean(episode_times[-verbose_freq:]):.3f}s | Rep. Mem. Size: {len(agent.replay_memory)}")

    minutes = int(np.sum(episode_times) // 60)
    seconds = np.sum(episode_times) % 60
    print(f"\nThe training process took {minutes} minutes and {seconds:.3f}s, with an average of {np.mean(episode_times):.3f}s, " + \
          f"and a max of {np.max(episode_times):.3f}s")

    print("-----------------------------------------------------------------------------------------------------------------------------------------")

    return (rewards, costs, episode_times, learning_times, steps, outcomes, rep_mem_sizes, epsilons), \
         (episode_times, env_step_times, action_taking_times, learning_times)
```

# Hyper-Parameters

```python
hyper_params = {
    "epsilon":               1,
    "epsilon_decay":         0.0008,
    "epsilon_min":           0.1,
    
    "gamma":                 0.99,
    "lr":                    0.001,
    
    "max_mem_size":          1_000_000,
    "minibatch_size":        512,
    "online_to_target_freq": 80,
    
    "nb_episodes":           2_000,
    "nb_steps":              200
}

hyper_params["verbose_freq"] = hyper_params["nb_episodes"] // 5
```

```python
# Mini Batch Size Is not Very important
hyper_param_names = ["nb_steps"]
hyper_param_values = [[100, 150, 200]]
```

```python
iterations = []
T = time.time()

for i, name in enumerate(hyper_param_names):
    for value in hyper_param_values[i]:
        hyper_params_copy = hyper_params.copy()
        
        hyper_params_copy[name] = value
        print(f"With {name} set to {value}.")

        agent = Agent(nb_states, nb_actions,
                      seed=          42,
                      max_mem_size=  hyper_params_copy["max_mem_size"],
                      minibatch_size=hyper_params_copy["minibatch_size"],
                      lr=            hyper_params_copy["lr"],
                      gamma=         hyper_params_copy["gamma"],
                      epsilon=       hyper_params_copy["epsilon"],
                      epsilon_decay= hyper_params_copy["epsilon_decay"],
                      epsilon_min=   hyper_params_copy["epsilon_min"])

        # (rewards, costs, episode_times, learning_times, steps, outcomes, rep_mem_sizes, epsilons), 
        #    (episode_times, env_step_times, action_taking_times, learning_times)
        data = train_agent(agent, hyper_params_copy)

        agent.save(f"models/{name}={value}.pth")
        
        iterations.append( data )

num_iterations = len(iterations)

training_time = time.time() - T
minutes = int(training_time // 60)
seconds = training_time % 60
print(f"\nThe whole training process took {minutes} minutes and {seconds:.3f}s.")
```

```python
rewards_list = []
costs_list = []
outcomes_list = []

for i, iteration in enumerate(iterations):
    rewards, costs, episode_times, learning_times, steps, outcomes, rep_mem_sizes, epsilons = iteration[0]

    rewards_list.append(  np.mean(rewards[-hyper_params["verbose_freq"]:])  )
    costs_list.append(    np.mean(rewards[-hyper_params["verbose_freq"]:])  )
    outcomes_list.append( np.mean(outcomes[-hyper_params["verbose_freq"]:]) )

# Rewards
if len(rewards_list) == len(set(rewards_list)):
    print(f"In Terms Of Rewards: Training {np.argmax(rewards_list)} Is the best with {rewards_list[np.argmax(rewards_list)]:.2f}")
else:
    print("All The Same max reward.")

# Costs
if len(costs_list) == len(set(costs_list)):
    print(f"In Terms Of Costs:   Training {np.argmin(costs_list)} Is the best with {costs_list[np.argmin(costs_list)]:.2f}")
else:
    print("All The Same min cost.")

# Wins
if len(outcomes_list) == len(set(outcomes_list)):
    print(f"In Terms Of Outcomes:    Training {np.argmax(outcomes_list)} Is the best with {outcomes_list[np.argmax(outcomes_list)]:.2f}")
else:
    print("All The Same max outcomes.")
```

# Visualize Some Parameters

```python
def scatter_data(data, color1, color2, padding, xlabel, ylabel, title, fig, grid, index):
    fig.add_subplot(*grid, index)
    
    plt.scatter( list(range(len(data))), data, color=color1 )

    if padding != 0:
        line = np.zeros(len(data))
        for i in range(padding, len(line)):
            line[i] = np.mean(data[i-padding: i+padding])
        line[: padding] = np.full( (padding, ), np.mean(data[: padding]) )
        
        plt.plot( list(range(len(line))), line, color=color2 )
    
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    plt.title(title)
```

```python
fig = plt.figure(figsize=(18, 16))
grid = (6, num_iterations)

for i, iteration in enumerate(iterations):
    rewards, costs, episode_times, learning_times, steps, outcomes, rep_mem_sizes, epsilons = iteration[0]
    
    scatter_data(rewards,        "C0", "C1", 100, "Episodes", "Rewards", "Evolution of Rewards by Episodes",             fig, grid, 0*num_iterations+i+1)
    scatter_data(costs,          "C1", "C2", 100, "Episodes", "Costs", "Evolution of Costs by Episodes",                 fig, grid, 1*num_iterations+i+1)
    scatter_data(episode_times,  "C2", "C3", 100, "Episodes", "Elapsed Time", "Evolution of Elapsed Time by Episodes",   fig, grid, 2*num_iterations+i+1)
    scatter_data(steps,          "C3", "C4", 100, "Episodes", "Steps", "Evolution of Steps by Episodes",                 fig, grid, 3*num_iterations+i+1)
    scatter_data(outcomes,       "C4", "C5", 0,   "Episodes", "Outcomes", "Evolution of Outcomes by Episodes",           fig, grid, 4*num_iterations+i+1)
    scatter_data(rep_mem_sizes,  "C5", "C6", 0,   "Episodes", "Rep Mem Sizes", "Evolution of Rep Mem Sizes by Episodes", fig, grid, 5*num_iterations+i+1)

plt.tight_layout()
plt.show()
```
